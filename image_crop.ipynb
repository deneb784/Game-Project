```python
from PIL import Image
import numpy as np

img = Image.open("raw_image\image_6.png").convert("RGB")
arr = np.array(img)

# 상단 여백 확인
top = 0
for i in range(arr.shape[0]):
    if np.sum(arr[i] != 255) > 0:  # 완전히 흰색이 아닌 첫 행 찾기
        top = i
        break

# 하단 여백 확인
bottom = arr.shape[0]
for i in range(arr.shape[0]-1, -1, -1):
    if np.sum(arr[i] != 255) > 0:
        bottom = i
        break

# 좌측 여백 확인
left = 0
for i in range(arr.shape[1]):
    if np.sum(arr[:,i] != 255) > 0:
        left = i
        break

# 우측 여백 확인
right = arr.shape[1]
for i in range(arr.shape[1]-1, -1, -1):
    if np.sum(arr[:,i] != 255) > 0:
        right = i
        break

print(f"상: {top}, 하: {arr.shape[0]-bottom-1}, 좌: {left}, 우: {arr.shape[1]-right-1}")
```

```python
from PIL import Image
import os

# ------------------- 설정 -------------------
input_folder = "raw_image"    # 원본 이미지 폴더
output_folder = "images"  # 저장 폴더
os.makedirs(output_folder, exist_ok=True)

top_crop = 21
bottom_crop = 21
left_crop = 26
right_crop = 26

rows, cols = 3, 3  # 격자 분할 수
# -------------------------------------------

count = 1  # 전체 이미지 순번

# 입력 폴더의 이미지 모두 처리
for file_name in os.listdir(input_folder):
    if file_name.lower().endswith((".png", ".jpg", ".jpeg")):
        img_path = os.path.join(input_folder, file_name)
        img = Image.open(img_path)

        # 여백 제거
        width, height = img.size
        cropped_img = img.crop((
            left_crop, 
            top_crop, 
            width - right_crop, 
            height - bottom_crop
        ))

        # 잘린 이미지 크기
        crop_width, crop_height = cropped_img.size
        cell_width = crop_width // cols
        cell_height = crop_height // rows

        # 순서대로 저장 (1,2,3...)
        for i in range(rows):
            for j in range(cols):
                left = j * cell_width
                upper = i * cell_height
                right = (j + 1) * cell_width
                lower = (i + 1) * cell_height

                cell = cropped_img.crop((left, upper, right, lower))
                cell_file = f"{count}.png"
                cell.save(os.path.join(output_folder, cell_file))
                count += 1

print("모든 작업 완료!")
```
